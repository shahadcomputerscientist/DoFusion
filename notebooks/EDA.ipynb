## Total value distribution

```python
df["total"].hist(bins=20)
plt.title("Receipt Total Distribution")
plt.show()
```

##### From the histogram we observe that the x-axis extends up to 8e7 (80,000,000), 
while the majority of receipt totals fall within the range [0–1]. 
This indicates the presence of extreme outliers in the dataset, likely caused 
by OCR extraction errors or abnormal receipt values. These outliers stretch 
the distribution and compress most observations near zero.
